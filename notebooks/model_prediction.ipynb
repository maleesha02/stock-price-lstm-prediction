```python
from pathlib import Path
import numpy as np
import joblib
from keras.models import load_model


def find_file(filename):
    """Search common project locations for a file."""
    possible_paths = [
        Path(filename),
        Path('notebooks') / filename,
        Path('..') / filename,
        Path('../notebooks') / filename,
    ]
    for path in possible_paths:
        if path.exists():
            return path
    raise FileNotFoundError(f"Could not find {filename}. Put it in notebooks/ or project root.")
```

```python

model_path = find_file('LSTM-Stock-Market.keras')
model = load_model(model_path)
print('Model loaded from:', model_path)
```

```python

scaler_path = find_file('stock-scaler.sav')
scaler = joblib.load(scaler_path)
print('Scaler loaded from:', scaler_path)
```

```python

input_prices = np.array([
    60.6428, 60.6928, 60.9085, 61.1571, 61.9599, 62.0214, 60.9999,
    60.2242, 60.0385, 57.8557, 55.4242, 56.0914, 57.7128, 56.2199,
    58.7464, 58.5442, 60.0642, 62.1571, 63.4942, 63.1114, 64.4728,
    65.1014, 66.4242, 65.5771, 65.6871, 65.4242, 64.5014, 64.8356,
    62.7371, 60.4628, 62.7214, 61.7014, 62.5928, 63.4357, 62.2785,
    62.9785, 64.2714, 62.8571, 63.6642, 64.6428, 64.3899, 64.7456,
    63.6642, 63.6385, 62.3571, 63.5328, 62.2485, 62.7857, 61.7857,
    62.1999
]).reshape(-1, 1)

print('Input shape before scaling:', input_prices.shape)
```

```python

input_scaled = scaler.transform(input_prices)
input_scaled = input_scaled.reshape(1, 50, 1)

print('Input shape for LSTM:', input_scaled.shape)
```

```python

predicted_open_price = model.predict(input_scaled)[0, 0]
print(f'Predicted next open price: {predicted_open_price:.4f}')
```
